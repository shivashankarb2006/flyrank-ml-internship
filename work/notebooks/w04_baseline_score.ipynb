## My Rule

These pages are more likely to benefit from content optimization.

## 2. Build the ranked queue (writes the CSV)

*Code the score, rank everything, write work/outputs/baseline_action_score.csv.*

```python
import pandas as pd
import os

df = pd.read_csv("flyrank-ml-internship-starter/data/raw/content_refresh_anonymized.csv")

score = (
    (df["ctr"] < 0.10).astype(int) * 30 +
    (df["avg_position"] > 20).astype(int) * 30 +
    (df["trend_direction"] == "down").astype(int) * 20 +
    (df["days_since_last_update"] > 180).astype(int) * 20
)

df["baseline_score"] = score

def reason(row):
    if row["ctr"] < 0.10:
        return "LOW_CTR"
    elif row["avg_position"] > 20:
        return "LOW_POSITION"
    elif row["trend_direction"] == "down":
        return "TRAFFIC_DECLINE"
    else:
        return "STALE_CONTENT"

df["reason_code"] = df.apply(reason, axis=1)

df["action"] = "Optimize Content"

queue = df.sort_values("baseline_score", ascending=False)

os.makedirs("work/outputs", exist_ok=True)

queue.to_csv("work/outputs/baseline_action_score.csv", index=False)

print(queue[[
    "content_id",
    "baseline_score",
    "reason_code",
    "action"
]].head(20))
```
